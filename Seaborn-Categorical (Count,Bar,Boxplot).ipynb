#### 1.1. Create a basic Countplot to get number of transatction for each of thr product line and the change the figure size

```python
sns.countplot(x="Product line",data=mart)
# Here the lable at x axis is overlapped  
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",data=mart)
```

```python
mart["Product line"].value_counts()
```

#### 1.2  Make it horizontal

```python
sns.countplot(y="Product line",data=mart)
```

#### 1.3. Add hue to get the count on two categies i.e Product and Gender

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",hue="Gender",data=mart)
plt.show()
```

#### 1.4.  using color palette

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",hue="Gender",data=mart,palette="crest")
plt.show()
```

```python
pd.crosstab(mart["Product line"],mart["Gender"])
```

##### change style using facecolor, linewidth, and edge color

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",
              facecolor=(0,0,0,0),data=mart,
              palette="crest",
             linewidth=5,
             edgecolor=sns.color_palette("dark",3))
plt.show()
```

#### 2.1. Creating Barplot

#### 2.2. Hue to barplot

#### 2.3. Horizontal Barplot

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Total",y="Product line",hue="Gender",data=mart)
```

#### 2.4. Order 

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",hue="Gender",data=mart)
```

```python
x=mart["Product line"].sort_values()
x.unique()
```

```python
# order, hue_order
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",hue="Gender",data=mart,
           order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'],
           hue_order=["Male","Female"])
```

#### 2.5. Add cap to ERROR Bar 

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",hue="Gender",data=mart,
           capsize=0.2)
```

#### 2.6. Remove the error bar using ci

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",hue="Gender",data=mart,ci=None)
```

#### 2.6. change bar color using COLOR attribute

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,ci=None,color="orange")
```

#### 2.7. Change color using pallete attribute 

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",hue="Gender",data=mart,ci=None,
           palette="Spectral_r")
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,ci=None,
           palette="Spectral_r")
```

#### 2.8 Saturation 

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,ci=None,
           palette="Spectral_r",saturation=0.5)
```

#### 2.9. Change default aggregation mathod using estimator parameter

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,ci=None,
           palette="Spectral_r",estimator=sum)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,ci=None,
           palette="Spectral_r",estimator=max)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,ci=None,
           palette="Spectral_r",estimator=min)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x="Product line",y="Total",data=mart,ci=None,
           palette="Spectral_r",estimator=np.median)
```

#### 3.5 Add MEAN marker in the box plot using Showmaens attribute and change its style using meanprops

```python
sns.boxplot(x="Payment",y="Total",hue="Gender",data=mart,showmeans=True
        , meanprops={"marker":"o","markerfacecolor":"white","markersize":5,"markeredgecolor":"black"})
```

#### 3.6 HORIZONTAL boxplot

```python
sns.boxplot(y="Payment",x="Total",hue="Gender",data=mart,showmeans=True
        , meanprops={"marker":"o","markerfacecolor":"white","markersize":5,"markeredgecolor":"black"})
```

#### 3.7. Change Palette,lineWidth

```python
sns.boxplot(y="Payment",x="Total",hue="Gender",data=mart,showmeans=True
        , meanprops={"marker":"o","markerfacecolor":"white","markersize":5,"markeredgecolor":"black"},
           palette="Spectral_r",linewidth=2)
```
